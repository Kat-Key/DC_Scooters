```python
import os
import pandas as pd
import json
filepath = 'C:\\Users\\Invario\\Desktop\\Data_Downloads_byCompany\\Bird'
bird_file=pd.DataFrame()
entries = os.listdir(filepath)
for entry in entries:
    with open(filepath + "\\" + entry) as data_file:    
        data_bike = json.load(data_file)  
        data = data_bike["data"]
        #print(bike)
        #bikes = data["bikes"]
        #print(bikes)
        df = pd.DataFrame.from_dict(data, orient='columns')
        #print(df)
        df2 = pd.DataFrame(df.bikes.tolist(), columns=['bike_id', 'lat', 'lon', 'battery_level',  'vehicle_type',  'reserved',  'disabled',])
        df2['filename'] = os.path.basename(entry)
        df2[['Month','Day','Time', 'Company']] = df2.filename.str.split("-",expand=True,)
    bird_file=bird_file.append(df2,ignore_index=True)

print(bird_file)
#bird_file.to_csv('bird_consolidated.csv')

 

        
    
```
